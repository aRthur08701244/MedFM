```python
import os
import pandas as pd
```

```python
dataset_files = os.listdir('./')
dataset_files = [i for i in dataset_files if i[-2:]=='py']
```

```python
name_list = []
num_line_list = []

for file in dataset_files:
    with open(file, 'r') as f:
        name_list.append(file)
        
        num_line = len(f.read().split('\n'))
        num_line_list.append(num_line)
        
        print(file, ": ", num_line)
```

```python
df_datasets = pd.DataFrame({"name":name_list , "num_line":num_line_list}).sort_values(by=['num_line']).reset_index(drop=True)
df_datasets
```
